Construisez une nouvelle DataFrame qui ne contient que les colonnes qui sont complètes à au moins 90%.

```python
# version longue
tokeep = []
for k in df:
  pna = df[k].count()/len(df)
  if pna > 0.9:
    tokeep.append(k)

rdf = df[tokeep]
rdf

# version courte
rdf = df.dropna(axis=1, thresh=0.9*len(df))
rdf
```

Dans le nouvelle DataFrame, trouvez pour chaque colonne le nombre de valeurs abhérantes en utilisant la méthode basée sur le z-score (>3). 

- Réduisez la DataFrame aux colonnes ayant moins de 5000 valeurs abhérantes. 
- Remplacez les valeurs abherantes par NaN dans toutes les autres.

```python
import numpy as np

crdf = rdf.copy()
for x in rdf: 
  if rdf[x].dtype != np.object:
    va = rdf[abs((rdf[x] - rdf[x].mean())/rdf[x].std()) > 3]
    if len(va) > 0 and len(va) < 5000:
       print(f"colonne {x} a {len(va)} valeurs abhérantes")
       crdf.loc[va.index, x] = np.nan
    else:
      print(f"drop {x}")
      crdf.drop(x, axis=1, inplace=True)

crdf.count()/rdf.count()
```

Créez deux versions de la DataFrame, une avec toutes les colonnes numériques normalisées en utilisant la normalisation minmax, et l'autre où ces colonnes sont standardisées (c'est-à-dire centrées sur la moyenne). 

```python
import numpy as np
numcols = []
for k in rdf:
  if rdf[k].dtype != np.object: 
    numcols.append(k)

ndf = (rdf[numcols]-rdf[numcols].min())/(rdf[numcols].max()-rdf[numcols].min())
ndf.describe()
```

```python
sdf = (rdf[numcols]-rdf[numcols].mean())/rdf[numcols].std()
sdf.describe()
```

Completez les valeurs manquantes dans toutes les colonnes numériques incomplètes dans ces deux DataFrames de la façon suivante : 
- Dans la DataFrame normalisée, remplir les valeurs manquantes avec la moyenne de la colonne correspondante.
- Dans la DataFrame standardisée, remplir chaque valeur manquante par la moyenne des valeurs correspondantes dans les 10 rangées les plus similaires, en utilisant toutes les colonnes numériques 100% complètes pour calculer la similarité/distance.


```python
# dataframe normalisée

for k in ndf : 
  if ndf[k].count() < len(ndf) and ndf[k].dtype != np.object:
    print(k,"before",ndf[k].count())
    ndf.loc[ndf[k].isnull(), k] = ndf[k].mean()
    print(k,"after",ndf[k].count())
```

```python
# dataframe standardisée

# trouver les plus proches voisins (les éléments les plus similaires)
# dans un grand volume de données est plutôt complex. On va donc 
# utiliser la librairie scikitlearn qui implemente des algorithmes 
# efficaces pour la recherche de plus proches voisins. 
#
# Attention : ce code peut prendre plusieurs minutes à s'executer
from sklearn.neighbors import NearestNeighbors

# on récupère la liste des colonnes numériques complètes
nccols = [] 
for k in sdf: 
  if sdf[k].dtype != np.object and sdf[k].count() == len(sdf): nccols.append(k)
print(nccols)

# on commence par créer l'objet dans lequel on va chercher les plus proche voisin
# et on lui donne comme données à utiliser la dataframe réduite aux colonnes
# numériques complètes.
nns = NearestNeighbors(n_neighbors=11, algorithm="kd_tree") # 11 car le point sur lequel on chercher
# sera dans la liste
# il y a plein de paramètres possibles. Voir la doc.
nns.fit(np.array(sdf[nccols]))

for k in sdf : 
  # on le fait seulement pour les colonne ayant plus de 97% de leurs valeurs renseignée (trop long sinon)
  if sdf[k].count() < len(sdf) and sdf[k].count()>0.97*len(sdf) and sdf[k].dtype != np.object:
    print(k,"before",sdf[k].count())
    for idx in sdf[sdf[k].isnull()].index:
      # moyennes des 10 plus proches voisins
      sdf.loc[idx, k] = sdf.loc[nns.kneighbors([sdf.loc[idx, nccols]], return_distance=False)[0], k].mean()
      # note that sometimes the 10 closests will have a nan value... we should normally take that into account
      print(sdf.loc[idx, k], end=" ")
    print("\n", k,"after",sdf[k].count())

```

Vous remarquerez que les valeurs sont des chaines de caractères. Transformez ces valeurs pour obtenir les valeurs numériques qui se trouve avant le caractère '[' dans chaque colonne.

Si `df` est une DataFrame, `df.plot()` affiche un graphe de ses colonnes numériques. Utilisez cette commande pour visualiser les valeurs dans les trois colonnes de la DataFrame.

```python
# correct to get the numeric values
def getNum(x):
  if type(x) == str : 
    return float(x[:x.index('[')-1])
  return x
rtodf ["France"] = rtodf["France"].apply(getNum)
rtodf ["Germany"] = rtodf["Germany"].apply(getNum)
rtodf ["Spain"] = rtodf["Spain"].apply(getNum)
rtodf.plot()
```
