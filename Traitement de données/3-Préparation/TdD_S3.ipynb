```python
df = pd.read_csv("bodies.csv")
nanvalue=float("NaN")
df.replace(" ", nanvalue, inplace=True)
df.replace("  ", nanvalue, inplace=True)
df.replace("   ", nanvalue, inplace=True)
df = df.rename(columns={"vitesse": "Speed", "hand.":"Handling"})
df["Speed"] = df.Speed.astype(np.dtype("float"))
df["Acceleration"] = df.Acceleration.astype(np.dtype("float"))
df["Weight"] = df.Weight.astype(np.dtype("float"))
df["Traction"] = df.Traction.astype(np.dtype("float"))
df["Handling"] = df.Handling.astype(np.dtype("float"))
def getSim(x):
  if x.Vehicle != df.loc[28].Vehicle:
    return abs(x.Speed - df.loc[28].Speed) + abs(x.Acceleration - df.loc[28].Acceleration) + abs(x.Weight - df.loc[28].Weight) + abs(x.Traction - df.loc[28].Traction)
sims = df.apply(getSim, axis=1)
df.loc[28,"Handling"] = df.loc[sims.idxmin()].Handling
df.loc[28]
```

```python
df = pd.read_csv("bodies.csv")
nanvalue=float("NaN")
df.replace(" ", nanvalue, inplace=True)
df.replace("  ", nanvalue, inplace=True)
df.replace("   ", nanvalue, inplace=True)
df = df.rename(columns={"vitesse": "Speed", "hand.":"Handling"})
df["Speed"] = df.Speed.astype(np.dtype("float"))
df["Acceleration"] = df.Acceleration.astype(np.dtype("float"))
df["Weight"] = df.Weight.astype(np.dtype("float"))
df["Traction"] = df.Traction.astype(np.dtype("float"))
df["Handling"] = df.Handling.astype(np.dtype("float"))
m = "Medium"
df["Category"] = df["Category"].replace("Meduim",m).replace("medium",m).replace("med",m).replace("Med",m).replace("small","Small")
df["Category"] = df["Category"].replace("Super Small", "Small")
df["Category"].value_counts()
```

```python
vals = ["Small", "Medium", "Large"]
def cat(x):
  if x != None: 
    try :
        return vals.index(x)
    except :
      return None
df["Category"] = df["Category"].apply(cat)
df
```

```python
df["type"] = "kart"
def isBike(x):
  return type(x) != float and ("Bike" in x or x == "Comet" or x=="Varmint" or x=="Mr. Scooty" or x=="Master Cycle")
df.loc[df.Vehicle.apply(isBike), "type"] = "bike"
df
```
