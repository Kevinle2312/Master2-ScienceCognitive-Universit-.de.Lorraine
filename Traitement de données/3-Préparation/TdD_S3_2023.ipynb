# Nettoyage

```python
import pandas as pd

df =  pd.read_csv("https://mdaquin.github.io/d/tdd/bodies.csv")
df
```

```python
df = df.rename(columns = {"vitesse": "Speed", "hand.": "Handling"})
df
```

```python
print(df[df["Vehicle"].isnull()])
print(df[df["Vehicle"]=="  "])
nanval = float("NaN")
df["Vehicle"] = df["Vehicle"].replace("  ", nanval)
df[df["Vehicle"].isnull()]
```

```python
print(df["Category"].value_counts())
m = "Medium"
df["Category"] = df["Category"].replace("Meduim",m).replace("medium",m).replace("med",m).replace("Med",m).replace("small","Small")
df["Category"].value_counts()
```

```python
df["Category"] = df["Category"].replace("Super Small", "Small")
df["Category"].value_counts()
```

```python
df.dtypes
# df["Speed"].mean()
```

```python
df["Speed"].apply(lambda x: float(x))
```

```python
nanvalue = float("NaN")
df.replace(" ", nanvalue, inplace=True)
df[df["Speed"] == " "]
```

```python
df["Speed"] = df["Speed"].apply(lambda x: float(x))
df["Speed"].dtypes
```

```python
for k in df:
  if k not in ["Vehicle", "Category"]: df[k] = df[k].apply(lambda x: float(x))
df.dtypes
```

```python
df[["Speed", "Acceleration", "Weight", "Handling", "Traction", "Mini Turbo"]] = df[["Speed", "Acceleration", "Weight", "Handling", "Traction", "Mini Turbo"]].astype(float)
df.dtypes
```

```python
df.fillna(0, inplace=True)
print(df[df.duplicated()])
df[df["Vehicle"] == "Sneeker"]
```

```python
df.fillna(0, inplace=True)
df.drop_duplicates(keep="first", inplace=True)
df[df.duplicated()]
df[df["Vehicle"] == "Sneeker"]
```

```python
df.describe()
```

```python
zscores = (df.Speed - df.Speed.mean()) / df.Speed.std()
df[abs(zscores) > 3]
```

```python
df[abs((df.Speed - df.Speed.mean()) / df.Speed.std()) > 3]
```

```python
vz = (df.Speed.value_counts() - df.Speed.value_counts().mean()) / df.Speed.value_counts().std()
print(f"rarest value is {vz.idxmin()}")
df[df.Speed == vz.idxmin()]
```
