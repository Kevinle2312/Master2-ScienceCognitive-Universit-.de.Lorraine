Vous remarquerez qu'il manque pas mal de valeurs pour la colonne `Win_Prob`. Créer une nouvelle DataFrame ne contenant que le match en question et complétez dans celui-ci les valeurs de `Win_Prob` selon trois méthodes :

Affichez les valeurs de `Win_Prob` complétées en utilisant ces trois méthodes dans un graphique.

1. En les remplaçants par la moyenne

```python
nfl_data_decmbre_31122027["WP_m1"] = nfl_data_decmbre_31122027.Win_Prob.fillna(nfl_data_decmbre_31122027.Win_Prob.mean()).plot()
```

2. En les remplaçants par la valeur de l'instant le plus similaire d'après `TimeSecs`

```python
nfl_data_decmbre_31122027.TimeSecs
```

```python
def getMinSims(x):
    sims = abs(nfl_data_decmbre_31122027.drop(x.name).TimeSecs - x.TimeSecs)
    return nfl_data_decmbre_31122027.loc[sims.idxmin(), "Win_Prob"]

nfl_data_decmbre_31122027["WP_m2"] = nfl_data_decmbre_31122027.loc[nfl_data_decmbre_31122027.Win_Prob.isnull(), ["Win_Prob", "TimeSecs"]].apply(getMinSims, axis=1)

nfl_data_decmbre_31122027.loc[nfl_data_decmbre_31122027.Win_Prob.isnull(), "WP_m2"].plot()
```

3. Par interpolation

```python
nfl_data_decmbre_31122027["WP_m3"] = nfl_data_decmbre_31122027["Win_Prob"].interpolate()
```

```python
nfl_data_decmbre_31122027["WP_m3"].plot()
```

**Bonus** : La fonction `.groupby("nom de colonne")` permet de regrouper les lignes d'une dataframe d'après les valeurs d'un colonne et d'y appliquer des fonctions d'aggregation (min, max, mean, count,...) Utilisez la pour visualiser le nombre de "moment" par année (colonne `Season`). Comment feriez vous pour visualiser le nombre de matchs (GameID)?

```python
from matplotlib import pyplot as plt

# Convert the 'Date' column to a datetime object
nfl_data_rr['Date'] = pd.to_datetime(nfl_data_rr['Date'])

# Group by year and calculate the mean of 'Value'
grouped = nfl_data_rr.groupby(nfl_data_rr['Date'].dt.year)['GameID'].count()

# Plot the data
plt.figure(figsize=(10, 6))
grouped.plot(kind='bar', color='skyblue')
plt.title('Count Match by Year')
plt.xlabel('Year')
plt.ylabel('Match')
plt.xticks(rotation=45)
plt.show()
```
